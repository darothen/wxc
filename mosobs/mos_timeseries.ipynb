```python
full_model_name = { 'NAM': "NAM-MET", "GFS": "GFS-MAV" }
months = {"JAN":1, "FEB":2, "MAR":3, "APR":4, "MAY":5, "JUNE":6,
          "JULY":7, "AUG":8, "SEPT":9, "OCT":10, "NOV":11,
          "DEC":12}
data_path = "data_arch/"

import datetime
import itertools
from util.mos import process_MOS

def concatenate_MOS(station, forecast_date, gfs=True, nam=True):
    """ Concatenate MOS forecasts from GFS and NAM corresponding to
    a given forecast date.
    
    Forecast period is implied to be 00Z on the forecast date to
    00Z on the following day. This method will align the 00Z-18Z 
    forecasts on the previous day to get a timeseries/snapshot of the
    forecast for the forecast date. 
    
    Example usage:
    
    >>> forecast_date = datetime.strptime("2008-09-17", "%Y-%m-%d")
    
    Parameters
    ----------
    station : string
        Station identifier code
    forecast_date : datetime.datetime
        The date corresponding to the 00z-00z forecast period
    gfs, nam : boolean, optional
        Logical switch on which model to include (both true by
        default)
    
    Returns
    -------
    pandas.DataFrame
        Aligned MOS forecasts
        
    """
    mos_pattern = "%s.%02d%02d%4d.%s.%02dZ" # station, mm, dd, yyyy, model, hh
    one_day = datetime.timedelta(days=1)
    
    ## Generate filenames of MOS data to read
    model_date = forecast_date - one_day
    md = model_date # alias
    
    hours = [0, 6, 12, 18]
    models = []
    if gfs: models.append("GFS")
    if nam: models.append("NAM")
    
    combos = itertools.product(hours, models)
    all_mos = {}
    for hour, model in combos:
        filename = mos_pattern % (station, md.month, md.day, md.year,
                                  full_model_name[model], hour)
        print "Processing", filename
        
        full_path = os.path.join(data_path, station, model, filename)
        if not os.path.exists(full_path):
            print "   could not find file; skipping"
            continue
            
        model_fcst_date = datetime.datetime(md.year, md.month, md.day, hour)
        with open(full_path, 'r') as f:
            proc_mos = process_MOS(f.readlines())
        
        fcst_name = "%s%02dZ" % (model, hour)
        all_mos[fcst_name] = proc_mos
    
    mos_panel = pd.Panel(all_mos)
    
    return mos_panel, all_mos
```

```python
STATION = "KAUS"
forecast_date = datetime.datetime.strptime("2010-07-21", "%Y-%m-%d")
model_date = forecast_date - datetime.timedelta(days=1)
fcst_begin = forecast_date 
fcst_end = fcst_begin + datetime.timedelta(days=1, hours=18)

fcst_bracket = [fcst_begin + datetime.timedelta(hours=6), 
                fcst_begin + datetime.timedelta(days=1, hours=6)]

mos_panel, all_mos = concatenate_MOS(STATION, forecast_date)
print "--"*36

print fcst_begin, fcst_end

fig = figure(figsize=(8,3))
ax = fig.add_subplot(111)
cc = ax._get_lines.color_cycle

## Plot temperatures and dewpoints
model_color = { 'NAM00Z': '#910000', 
                'NAM12Z': '#ED2B2B',
                'GFS00Z': '#090CB5',
                'GFS06Z': '#144BE3',
                'GFS12Z': '#256BE6',
                'GFS18Z': '#2D94ED', }
for key, color in zip(mos_panel, cc):
    raw_mos = all_mos[key]
    maxmins = raw_mos.maxmins[forecast_date]
    
    data = mos_panel[key].ix[fcst_begin:fcst_end]
    
    ## Master panel
    c = model_color[key]
    tp = data["TMP"].plot(ax=ax, style="o-", c=c, label=key)
    
    ## Single panel
    fig_solo = figure(figsize=(8,3))
    ax_s = fig_solo.add_subplot(111)
    tp_s = data["TMP"].plot(ax=ax_s, style="o-r")
    dp_s = data["DPT"].plot(ax=ax_s, style="x-g")
    
    ylo, yhi = ax_s.get_ylim()
    
    #ax_s.set_ylim(65, 90)
    ax_s.vlines(fcst_bracket, ylo, yhi, 
                linestyle='dashed', linewidth=3)
    ax_s.hlines([maxmins['max'], maxmins['min']], fcst_begin, fcst_end,
                linestyle='-.', colors=['r', 'b'])
    ax_s.text(fcst_begin, maxmins['max']-1.2, "%2dF" % maxmins['max'], color='r')
    ax_s.text(fcst_begin, maxmins['min']-1.2, "%2dF" % maxmins['min'], color='b')
    
    ax_s.set_title("Surface Temperature and Dewpoint (F), MOS, %s %s %s" %
                   (STATION, key, model_date.strftime("%m/%d/%y")))
    
ax.legend(fontsize=10)
ax.set_xlim(fcst_bracket)
draw()

ylo, yhi = ax.get_ylim()
#ax.set_ylim(65, 90)

show()
```

### MOS Meteograms

```python
import matplotlib.dates as mdates
from matplotlib.ticker import FuncFormatter

cloudy = {
    "CL": "#FFFFFF",
    "FW": "#E0E0E0",
    "SC": "#C0C0C0",
    "BK": "#A0A0A0",
    "OV": "#808080",
}
qpfs = ["no precip", "0.01-0.09", "0.10-0.24", "0.25-0.49",
        "0.50-0.99", "1.00-1.99", "$>$ 2.0"]

key = "GFS18Z"

rc('axes', facecolor='none')
#fig, [ax_top, ax_bot] = subplots(3, 1, figsize=(12, 5), sharex=True)
fig = figure(figsize=(12, 7))
ax_top = subplot2grid((5, 3), (0, 0), colspan=3, rowspan=2)
ax_mid = subplot2grid((5, 3), (2, 0), colspan=3, rowspan=1, sharex=ax_top)
ax_bot = subplot2grid((5, 3), (3, 0), colspan=3, rowspan=2, sharex=ax_top)

## Fine tuning
subplots_adjust(hspace=0.01)

raw_mos = all_mos[key]
d = raw_mos.ix[fcst_begin:]
maxmins = raw_mos.maxmins[forecast_date]

## Top panel - temperatures
tp = d["TMP"].plot(ax=ax_top, style="o-r", x_compat=True)
dp = d["DPT"].plot(ax=ax_top, style="x-g", x_compat=True)
ylo, yhi = ax_top.get_ylim()
ylo -=5 ; yhi += 5

ax_top.hlines([maxmins['max'], maxmins['min']], fcst_begin, fcst_end,
              linestyle='-.', colors=['r', 'b'])
ax_top.text(fcst_begin, maxmins['max']-1.2, "%2dF" % maxmins['max'], color='r')
ax_top.text(fcst_begin, maxmins['min']-1.2, "%2dF" % maxmins['min'], color='b')
ax_top.set_ylabel("Temp/Dewpoint (F)")


## Top panel - cloudiness
# calculate edges
times = map(lambda x: x.to_datetime(), d.index)

for i in range(len(times) - 1):
    if i == 0:
        c, r = times[i], times[i+1]
        l = c - (r-c)
    elif i == len(times)-1:
        l, c = times[i-1], times[i]
        r = c + (c-l)
    else:
        l, c, r = times[i-1], times[i], times[i+1]
    bnd_left = c - (c-l)/2
    bnd_right = c + (r-c)/2
    width = bnd_right - bnd_left
    
    #print bnd_left, bnd_right, d["CLD"].ix[i]
    
    c = cloudy[d["CLD"].ix[i].strip()]
    ax_top.axvspan(bnd_left, bnd_right, facecolor=c, alpha=0.5)

## Middle panel - wind barbs
dirs = np.pi*d["WDR"]*10./180. # in radians
speeds = d["WSP"] # 10m wind speed in knots  
us = -speeds.values*np.sin(dirs.values)
vs = -speeds.values*np.cos(dirs.values)

ax_mid.barbs(mdates.date2num(times), np.zeros_like(times), us, vs)
ax_mid.set_ylim(-1, 1)
ax_mid.set_ylabel("Winds")
ax_mid.axis('off')

## Bottom Panel - PoPs and QPF
# 6 hour
ax_qpf = ax_bot.twinx()
six_hour = d[['P06', 'Q06']].dropna()

times = map(lambda x: x.to_datetime(), six_hour.index)
dt = datetime.timedelta(hours=6)

rights = mdates.date2num(times)
lefts = mdates.date2num([t-dt for t in times])
ax_bot.hlines(six_hour['P06'].values, lefts, rights,
              linestyle='dashed', color='k')
ax_qpf.bar(lefts, six_hour['Q06'].values, rights-lefts,
            facecolor='g', alpha=0.4, hatch="/", edgecolor='k')
ax_qpf.grid(False)

# 12 hour
twelve_hour = d[['P12', 'Q12']].dropna()

times = map(lambda x: x.to_datetime(), twelve_hour.index)
dt = datetime.timedelta(hours=12)

rights = mdates.date2num(times)
lefts = mdates.date2num([t-dt for t in times])
ax_bot.hlines(twelve_hour['P12'].values, lefts, rights,
              linestyle='dashed', color='m')
ax_qpf.bar(lefts, twelve_hour['Q12'].values, rights-lefts,
            facecolor='g', alpha=0.4, hatch="\\", edgecolor='m')
ax_qpf.grid(False)

ax_bot.set_ylabel("PoP / QPF")
to_pct = FuncFormatter(lambda x, pos=0: "%3d" % x + "$\%$")
ax_bot.yaxis.set_major_formatter(to_pct)


## Clean up
## Fix ylims - all
ax_top.set_ylim(ylo, yhi)
ax_bot.set_ylim(0, 100.)
ax_qpf.set_ylim(0, 6.5)

# Convert tick format on bottom axis
ax_bot.xaxis.set_major_locator(mdates.DayLocator(interval=1))
ax_bot.xaxis.set_major_formatter(mdates.DateFormatter("%HZ\n%m/%d"))
ax_bot.xaxis.set_minor_locator(mdates.HourLocator(interval=3))
ax_bot.xaxis.set_minor_formatter(mdates.DateFormatter("%HZ"))

# Remove redundant tick labels
setp(ax_top.get_xticklabels(), visible=False)
setp(ax_top.get_xticklabels('minor'), visible=False)
setp(ax_mid.get_xticklabels(), visible=False)
setp(ax_mid.get_xticklabels('minor'), visible=False)

# Day 1 bounds
ax_top.vlines(fcst_bracket, ylo, yhi, 
              linestyle='dashed', linewidth=3)
ax_bot.vlines(fcst_bracket, 0, 100, 
              linestyle='dashed', linewidth=3)

# Re-label QPF
ax_qpf.set_yticklabels(qpfs)

# Adjust x limits
ax_bot.set_xlim(fcst_begin, fcst_end)
```
