```python
fig = figure(figsize=(9, 5))
ax = fig.add_subplot(111)

subset = core[(core.index.month == 10)]
subset['TMIN'].hist(label='TMIN', bins=12, normed=True, alpha=0.7)
subset['TMAX'].hist(label='TMAX', bins=12, normed=True, alpha=0.7)
ax.legend()
```
