### Temperature bias plots - MOS

```python
## Collect the observation and MOS data
from util.mos import concatenate_MOS

variable_data, obs = parse_observations(station)

date_begin_str = "2012-10-10"
date_end_str   = "2012-10-24"
dates = pd.date_range(date_begin_str, date_end_str)

all_obs = obs.ix[dates]
fcst_results = []
for d in dates:
    mos_panel, mos_fcst = concatenate_MOS(station, d)
    
    ## Select GFS18Z max/mins
    gfs18z = mos_fcst['GFS18Z']
    gfs18z_maxmins = gfs18z.maxmins[d]
    
    ## Select GFS 06Z max/mins
    gfs06z = mos_fcst['GFS06Z']
    gfs06z_maxmins = gfs06z.maxmins[d]
    
    fcst_date = [
        gfs06z_maxmins['max'], # GFS06Z_MAX
        gfs06z_maxmins['min'], # GFS06Z_MIN
        gfs18z_maxmins['max'], # GFS18Z_MAX
        gfs18z_maxmins['min'], # GFS18Z_MIN
        np.floor(all_obs['TMAX'].ix[d]), # OBS_MAX
        np.floor(all_obs['TMIN'].ix[d]), # OBS_MIN
    ]
    print d, obs, fcst_date
    fcst_results.append(fcst_date)

all_fcsts = pd.DataFrame(fcst_results, index=dates,
                         columns=["GFS06Z_MAX", "GFS06Z_MIN",
                                  "GFS18Z_MAX", "GFS18Z_MIN",
                                  "OBS_MAX", "OBS_MIN"])
print all_fcsts
```

```python
import matplotlib.dates as mdates

fig, [ax_hi, ax_lo] = subplots(2, 1, figsize=(16, 8), sharex=True)
subplots_adjust(hspace=0)

doff = datetime.timedelta(hours=6)

n_models = 2
mos_a = "GFS06Z"
mos_b = "GFS18Z"

glyphs = {
   #[marker, color_max, color_min]. Obs first, MOS iterations follow
    "OBS": ["", "#FF2424", "#0522FF"],
    mos_a: ["s", "#960303", "#061BBD"],
    mos_b: ["d", "#FF7373", "#818FFC"],
}

## Plot TMAX bias
gfs06z_bias = all_fcsts['%s_MAX' % mos_a] - all_fcsts['OBS_MAX']
gfs18z_bias = all_fcsts['%s_MAX' % mos_b] - all_fcsts['OBS_MAX']

dates_plt = mdates.date2num(dates)
zeroes = np.zeros_like(dates_plt)

ax_hi.plot_date(dates_plt, zeroes, marker=glyphs["OBS"][0],
                color=glyphs["OBS"][1], 
                lw=None, ms=10, mew=4)
for d, val in all_fcsts['OBS_MAX'].iteritems():
    ax_hi.text(d + doff, 0., "%02d" % val,
               {'size':16, 'weight': 'bold', 
                'color': glyphs['OBS'][1], 'family': 'sans-serif'},
               va='center')
    
## MOS A
ax_hi.plot(dates_plt, gfs06z_bias.values, label=mos_a,
           marker=glyphs[mos_a][0], color=glyphs[mos_a][1], 
           lw=0, ms=10, mfc=None)
for d, val in gfs06z_bias.iteritems():
    bias = val
    true_val = all_fcsts["%s_MAX" % mos_a].ix[d] 
    ax_hi.text(d + doff, bias, "%02d" % true_val,
               {'size':16, 'weight': 'regular',
                'color': 'k', 'family': 'sans-serif'},
               va='center')

## MOS B
ax_hi.plot(dates_plt, gfs18z_bias.values, label=mos_b,
           marker=glyphs[mos_b][0], color=glyphs[mos_b][1], 
           lw=0, ms=10, mfc=None)
for d, val in gfs18z_bias.iteritems():
    bias = val
    true_val = all_fcsts["%s_MAX" % mos_b].ix[d] 
    ax_hi.text(d + doff, bias, "%02d" % true_val,
               {'size':16, 'weight': 'regular',
                'color': 'k', 'family': 'sans-serif'},
               va='center')

ax_hi.set_ylabel("Max Temp Bias ($^o$F)")

#####################################################

## Plot TMIN bias
gfs06z_bias = all_fcsts['%s_MIN' % mos_a] - all_fcsts['OBS_MIN']
gfs18z_bias = all_fcsts['%s_MIN' % mos_b] - all_fcsts['OBS_MIN']

ax_lo.plot_date(dates_plt, zeroes, marker=glyphs["OBS"][0],
                color=glyphs["OBS"][2], lw=None, ms=10, mew=4)
for d, val in all_fcsts['OBS_MIN'].iteritems():
    ax_lo.text(d + doff, 0., "%02d" % val,
               {'size':16, 'weight': 'bold', 
                'color': glyphs['OBS'][2], 'family': 'sans-serif'},
               va='center')

## MOS A
ax_lo.plot(dates_plt, gfs06z_bias.values, label=mos_a,
           marker=glyphs[mos_a][0], color=glyphs[mos_a][2],
           lw=0, ms=10, mfc=None)
for d, val in gfs06z_bias.iteritems():
    bias = val
    true_val = all_fcsts["%s_MIN" % mos_a].ix[d] 
    ax_lo.text(d + doff, bias, "%02d" % true_val,
               {'size':16, 'weight': 'regular',
                'color': 'k', 'family': 'sans-serif'},
               va='center')
    
## MOS B
ax_lo.plot(dates_plt, gfs18z_bias.values, label=mos_b,
           marker=glyphs[mos_b][0], color=glyphs[mos_b][2],
           lw=0, ms=10, mfc=None)
for d, val in gfs18z_bias.iteritems():
    bias = val
    true_val = all_fcsts["%s_MIN" % mos_b].ix[d] 
    ax_lo.text(d + doff, bias, "%02d" % true_val,
               {'size':16, 'weight': 'regular',
                'color': 'k', 'family': 'sans-serif'},
               va='center')

ax_lo.set_ylabel("Min Temp Bias ($^o$F)")

############################

## Plot X/Y limits
ax_hi.set_xlim(dates_plt[0]-.5, dates_plt[-1]+.75)
ax_hi.set_ylim(ax_hi.get_ylim()[0]-1, ax_hi.get_ylim()[-1]+1)
ax_lo.set_ylim(ax_lo.get_ylim()[0]-1, ax_lo.get_ylim()[-1]+1)

## Format date ticker along bottom
ax_lo.xaxis.set_major_locator(mdates.DayLocator(interval=1))
ax_lo.xaxis.set_major_formatter(mdates.DateFormatter("%d\n%b\n%Y"))

## Label/color-key on RHS
ncol = int(math.ceil(n_models/2.))
ax_hi.legend(loc='best', ncol=ncol)
ax_lo.legend(loc='best', ncol=ncol)


ax_hi.set_title("MOS Temperature errors - %s" % station)

draw()
```
